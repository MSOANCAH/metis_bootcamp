````python
# Source: http://web.mta.info/developers/turnstile.html
def get_data(week_nums):
    url = "http://web.mta.info/developers/data/nyct/turnstile/turnstile_{}.txt"
    dfs = []
    for week_num in week_nums:
        file_url = url.format(week_num)
        dfs.append(pd.read_csv(file_url))
    return pd.concat(dfs)
        
week_nums = [170812, 170819, 170826]
turnstiles_df = get_data(week_nums)
````

Exercise 1.3

* These counts are for every n hours. (What is n?) We want total daily exits.

Now make it that we again have the same keys, but now we have a single value for a single day, which is the total number of passengers that entered through this turnstile on this day.


````python
turnstiles_daily = turnstiles_df.groupby(["C/A", "UNIT", "SCP", "STATION", "DATE"]).EXITS.first().reset_index()
````

````python
turnstiles_daily.head()
````

````python
turnstiles_daily[["PREV_DATE", "PREV_EXITS"]] = (turnstiles_daily
                                                       .groupby(["C/A", "UNIT", "SCP", "STATION"])["DATE", "EXITS"]
                                                       .transform(lambda grp: grp.shift(1)))
````

````python
turnstiles_daily.head(10)
````

````python
# Drop the rows for the first date
turnstiles_daily.dropna(subset=["PREV_DATE"], axis=0, inplace=True)
````

````python
turnstiles_daily[turnstiles_daily["EXITS"] < turnstiles_daily["PREV_EXITS"]].head(10)
````

The previous line indicates that some of the counters are operating in reverse.  Needs to be fixed!

Let's check out which stations are having this problem.

````python
turnstiles_daily[turnstiles_daily["EXITS"] < turnstiles_daily["PREV_EXITS"]].groupby(["C/A", "UNIT", "SCP", "STATION"]).size()
````

````python
def get_daily_counts(row, max_counter):
    counter = row["EXITS"] - row["PREV_EXITS"]
    if counter < 0:
        # May be counter is reversed?
        counter = -counter
    if counter > max_counter:
        #print(row["ENTRIES"], row["PREV_ENTRIES"])
        counter = min(row["EXITS"], row["PREV_EXITS"])
    if counter > max_counter:
        # Check it again to make sure we are not giving a counter that's too big
        return 0
    return counter
````

````python
# If counter is > 1Million, then the counter might have been reset.  
# Just set it to zero as different counters have different cycle limits
turnstiles_daily["DAILY_EXITS"] = turnstiles_daily.apply(get_daily_counts, axis=1, max_counter=1000000)
````

````python
turnstiles_daily = turnstiles_daily.drop(["EXITS", "PREV_EXITS","PREV_DATE"], axis=1, errors="ignore")
plt.figure(figsize=(8,6))
turnstiles_daily.groupby(["C/A", "UNIT", "SCP", "STATION"])['DAILY_EXITS'].sum().nlargest(10).plot(kind='barh')
top_ten_station = turnstiles_daily.groupby(["C/A", "UNIT", "SCP", "STATION"])['DAILY_EXITS'].sum().nlargest(10).reset_index()
ca_list = top_ten_station["C/A"].tolist()
unit_list = top_ten_station["UNIT"].tolist()
scp_list = top_ten_station["SCP"].tolist()
station_list = top_ten_station["STATION"].tolist()
mask = ((turnstiles_daily["C/A"].isin(ca_list)) & 
        (turnstiles_daily["UNIT"].isin(unit_list))& 
        (turnstiles_daily["SCP"].isin(scp_list)) & 
        (turnstiles_daily["STATION"].isin(station_list)))
top_10_exits = turnstiles_daily[mask].head(10)
````

````python
turnstiles_hourly_df = turnstiles_df.groupby(["C/A","UNIT","SCP","STATION"])["EXITS"].diff()
````

````python
turnstiles_hourly = turnstiles_df.groupby(["C/A", "UNIT", "SCP", "STATION", "DATE_TIME"]).EXITS.first().reset_index()
turnstiles_hourly[["PREV_DATE_DATETIME", "PREV_EXITS"]] = (turnstiles_hourly
                                                       .groupby(["C/A", "UNIT", "SCP", "STATION"])["DATE_TIME", "EXITS"]
                                                       .transform(lambda grp: grp.shift(1)))

turnstiles_hourly.dropna(subset=["PREV_EXITS"], axis=0, inplace=True)

def get_hourly_counts(row, max_counter):
    counter = row["EXITS"] - row["PREV_EXITS"]
    if counter < 0:
        # May be counter is reversed?
        counter = -counter

    if counter > max_counter:
#        print(row["EXITS"], row["PREV_EXITS"])
        counter = min(row["EXITS"], row["PREV_EXITS"])

    if counter > max_counter:
        # Check it again to make sure we are not giving a counter that's too big
        return np.nan

    return counter

turnstiles_hourly["HOURLY_EXITS"] = turnstiles_hourly.apply(get_hourly_counts, axis=1, max_counter=1000000)
````

````python
turnstiles_hourly
````

````python
def get_delta_time_in_number_of_hours(row): 

    dt = row["DATE_TIME"] 
    pdt = row["PREV_DATE_DATETIME"]
    if (max(pdt,dt) == pdt): 
        (dt,pdt) = (pdt,dt)
    delta_time = dt-pdt

    hours = delta_time.seconds//3600

    return hours

turnstiles_hourly["DELTA_HOURS"] = turnstiles_hourly.apply(get_delta_time_in_number_of_hours, axis=1)
````
