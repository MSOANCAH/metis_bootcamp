Consider the function f(x, y) = (x^2-8x+7)/(3) + (y^2-5x+8)/(6).

    * Analytically (with pen and paper!), calculate the gradient vector in terms of x and y
    * Use matplotlib to plot this function as a 2-D surface and get a feel for where its minimum value is

    * Pick a learning rate
    * Write code that starts at your starting point (x, y), then moves by your learning rate in the direction opposite the gradient (by plugging x and y into your calculated gradient)
    * Repeat until convergence and the minimum is found.

How do you know when the minimum is found aka when to stop? Up to you.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-30,30,num=1000)
y = np.linspace(-30,30,num=1000)
```

```python
xv, yv = np.meshgrid(x, y)
```

```python
f = (xv**2-8*xv+7)/3 + (yv**2-5*xv+8)/6
```

```python
plt.figure(figsize=(10,8))
plt.contour(xv,yv,f)
```

```python
point = [10,0]

def gradient_descent(point,alpha,steps):
    x=point[0]
    y=point[1]
    for i in range(steps):
        dx = (2*x/3-7/2)
        dy = y/3
        x = x-alpha*dx
        y = y-alpha*dy
    return([x, y])
```

```python
gradient_descent(point,.01,2000)
```
