### Challenge 1

Build a linear model that uses only a constant term (a column of ones) to predict a continuous outcome (like domestic total gross). How can you interpret the results of this model? What does it predict? Make a plot of predictions against actual outcome. Make a histogram of residuals. How are the residuals distributed?

First I will load in my house prices data set that I used for my Luther project.

```python
df = pd.read_pickle('Luther/Zillow_df_ver2.pkl')
```

```python
df.head()
```

```python
df['constant'] = 1
df.head()
```

```python
lr = LinearRegression()
x = df['constant'].values.reshape(-1,1)
y = df['sale_price'].values.reshape(-1,1)
lr.fit(x,y)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
fig_baseline = plt.figure(figsize=(10,8))
plt.plot(y,lr.predict(x),'o',label='model')
#x_dat = np.linspace(0,4)
#line_best_fit = lr.intercept_ + np.zeros(x_dat.shape)
x_dat = np.arange(9e5,4e6)
plt.plot(x_dat, x_dat, label='ideal')
plt.xlabel('Actual price ($)',fontsize=16)
plt.ylabel('Predicted price ($)', fontsize=16)
plt.title('Linear Regression Model With A Constant Feature',fontsize=16)
plt.legend()
```

The results of our model is a constant price.  This makes sense, because we are running a model for house prices using a constant feature.  The constant value that we predict is $1,308,650.

```python
lr
```

```python
plt.figure(figsize=(10,8))
plt.hist(lr.predict(x)-y,bins=50)
plt.xlabel('residuals',fontsize=14)
plt.ylabel('Number of houses',fontsize=14)
```

The residuals are distributed somewhat Normally, but are skewed on the left side (e.g. the predicted value is less than the actual)

### Challenge 2

Repeat the process of challenge one, but also add one continuous (numeric) predictor variable. Also add plots of model prediction against your feature variable and residuals against feature variable. How can you interpret what's happening in the model?

My predictor variable will be floor_size, which is strongly correlated with house price. 

```python
lr2 = LinearRegression()
X = df[['constant','floor_size']].values
y = df['sale_price'].values.reshape(-1,1)
lr2.fit(X,y)
```

```python
lr2.coef_[0,1]
```

```python
plt.figure(figsize=(10,8))
plt.plot(X[:,1],y,'o')
x_dat=np.arange(500,9000)
line_best_fit = lr2.intercept_ + lr2.coef_[0,1] * x_dat
plt.plot(x_dat, line_best_fit,'r-')
plt.xlabel('Floor size',fontsize=16)
plt.ylabel('House price ($)', fontsize=16)
plt.title('Linear Regression Model with 2 Features', fontsize=16)
```

```python
y2_predict = lr2.predict(X)
```

```python
plt.figure(figsize=(10,8))
plt.plot(X[:,1],(y2_predict-y),'o')
plt.xlabel('Floor size',fontsize=16)
plt.ylabel('Residuals', fontsize=16)
plt.title('Linear Regression Model with 2 Features', fontsize=16)
```

The plot indicates that the residuals are more or less normally distributed about 0.  This is good, because it means that our model is not biased in any particular way.  In addition, we can say that the variances are not heteroscedastic, since the variances are evenly distributed across all floor sizes.

### Challenge 3

Repeat the process of challenge 1, but add a categorical feature (like genre). You'll have to convert a column of text into a number of numerical columns ("dummy variables"). How can you interpret what's happening in the model?

I don't have any categorical features in my dataset, so I've decided to create artifical categorical features by converting # of bedrooms from ordinal to categorical.

```python
df['bedrooms_str'] = df['bedrooms'].astype(str)
```

```python
df['bedrooms_str'].dtype
```

```python
data_dummies = pd.get_dummies(df)
```

```python
data_dummies.head()
```

```python
lr3 = LinearRegression()
X = data_dummies[['constant','bedrooms_str_2','bedrooms_str_3','bedrooms_str_4','bedrooms_str_5','bedrooms_str_6']].values
y = data_dummies['sale_price'].values.reshape(-1,1)
lr3.fit(X,y)
```

```python
plt.figure(figsize=(10,8))
y_pred = lr3.predict(X)
plt.plot(y,y_pred,'o')
x_dat=np.arange(500,9000)
plt.xlabel('Actual price ($)',fontsize=16)
plt.ylabel('Predicted price ($)', fontsize=16)
plt.title('Linear Regression Model with categorical features', fontsize=16)
```

```python
plt.figure(figsize=(10,8))
plt.plot(y_pred,(y_pred-y),'o')
plt.xlabel('Predicted price ($)',fontsize=16)
plt.ylabel('Residuals', fontsize=16)
plt.title('Linear Regression Model with 2 Features', fontsize=16)
```

The results of this model are highly stratified, which makes sense, because the number of bedrooms is not very predictive of house prices.

### Challenge 5

Fitting and checking predictions on the exact same data set can be misleading. Divide your data into two sets: a training and a test set (roughly 75% training, 25% test is a fine split). Fit a model on the training set, check the predictions (by plotting versus actual values) in the test set.

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['sale_price']  
X = df[['floor_size','bathrooms','lot_size']] 
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
lr5 = LinearRegression()
lr5.fit(X_train,y_train)
```

```python
print("Training set score: {:.2f}".format(lr5.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lr5.score(X_test, y_test)))
```

```python
plt.figure(figsize=(10,8))
y_test_pred = lr5.predict(X_test)
plt.plot(y_test,y_test_pred,'o')
plt.xlabel('Actual price ($)',fontsize=16)
plt.ylabel('Predicted price ($)', fontsize=16)
plt.title('Linear Regression Model after Train Test Split', fontsize=16)
```

```python
plt.figure(figsize=(10,8))
y_test_pred = lr5.predict(X_test)
plt.plot(y_test_pred,(y_test_pred-y_test),'o')
plt.xlabel('Predicted price ($)',fontsize=16)
plt.ylabel('Residuals', fontsize=16)
plt.title('Linear Regression Model after Train Test Split', fontsize=16)
```
