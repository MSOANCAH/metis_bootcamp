### Challenge 1

Plot domestic total gross over time.

The first step is that we need to load in the data.

```python
movies_df = pd.read_csv('2013_movies.csv')
```

```python
movies_df.head()
```

```python
movies_df['ReleaseDate'].dtype
```

The above shows that the ReleaseDate column is of type Object, which in this case is a String.  So, we should convert it to a datetime object.

```python
movies_df["ReleaseDateTime"] = pd.to_datetime(movies_df['ReleaseDate'] , format="%Y-%m-%d %H:%M:%S")
```

```python
movies_df['ReleaseDateTime'].dtype
```

```python
movies_df.drop('ReleaseDate',axis=1,inplace=True)
movies_df.head()
```

Since the 'ReleaseDateTime' column is not sorted, we need to do that before plotting the data.

```python
movies_df = movies_df.sort_values('ReleaseDateTime',ascending=True)
```

```python
import seaborn as sns
```

```python
sns.set_style('whitegrid')
```

```python
plt.figure(figsize=(12,8))
plt.plot(movies_df['ReleaseDateTime'],movies_df['DomesticTotalGross'])
plt.ylabel('Domestic Total Gross ($)', fontsize=14)
plt.xlabel('Film Release Date', fontsize=14)
```

### Challenge 5

What director in your dataset has the highest gross per movie?

```python
movies_df['Director'].value_counts().sort_values(ascending=False).nlargest(10)
```

Since 3 of the directors had more than one movie, I will have to get the average for these directors.  I can do this using groupby.

### Challenge 6

Bin your dataset into months and make a bar graph of the mean domestic total gross by month. Error bars will represent the standard error of the mean.

Title of graph should include: Mean Domestic Total Gross by Month in 2013

Topic for consideration: what is the correct formula for the standard error of the mean? Examine the error bars and see if they are "reasonable."

```python
movies_df.head()
```

```python
import datetime as dt

A = dt.datetime.now()
A.month
```

```python
movies_df.loc[83,'ReleaseDateTime'].month
```

The following code extracts the month from the datetime object.

```python
movies_df['ReleaseMonth'] = movies_df['ReleaseDateTime'].apply(lambda row: row.month)
```

```python
movies_df.head()
```

Now we will do a groupby based on ReleaseMonth.

```python
movie_month_means = movies_df.groupby(['ReleaseMonth'])['DomesticTotalGross'].mean()
movie_month_means_stderr = movies_df.groupby(['ReleaseMonth'])['DomesticTotalGross'].sem()  #handy groupby method to calculate standard error of the mean
```

```python
plt.figure(figsize=(10,8))
plt.bar(range(1,13), movie_month_means, yerr=movie_month_means_stderr)
plt.xticks(range(1,13))
plt.xlabel('Month',fontsize=14)
plt.ylabel
plt.title('Mean Domestic Total Gross by Month in 2013', fontsize=14)
```
